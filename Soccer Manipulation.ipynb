```python
# This is to select most valuable players and low income players
df1 = pd.DataFrame(df, columns=["Name", "Wage", "Value"])
df1
```

```python
def value_to_float(x):
    if type(x) == float or type(x) == int:
        return x
    if 'K' in x:
        if len(x) > 1:
            return float(x.replace('K', '')) * 1000
        return 1000.0
    if 'M' in x:
        if len(x) > 1:
            return float(x.replace('M', '')) * 1000000
        return 1000000.0
    if 'B' in x:
        return float(x.replace('B', '')) * 1000000000
    return 0.0

wage = df1['Wage'].replace('[\€,]', '', regex=True).apply(value_to_float)
value = df1['Value'].replace('[\€,]', '', regex=True).apply(value_to_float)

df1['Wage'] = wage
df1['Value'] = value

df1['difference'] = df1['Value'] - df1['Wage']
df1
```

```python
df1.sort_values('difference', ascending=False)
```

```python
import seaborn as sns # Give more visualisation to data
sns.set() # This set or call on seaborn

graphs = sns.scatterplot(x='Wage', y='Value', data=df1) # var that hold scatterplot as the graphical method 
```

```python
from bokeh.plotting import figure, show # This lib allow us to interact with the graph method
from bokeh.models import HoverTool

TOOLTIPS= HoverTool(tooltips=[
    ("index", "$index"),
    ("(Wage,Value)", "(@Wage, @Value)"),
    ("Name", "@Name"),
]) # This method keeps my graph in checked
p = figure(title='Fifa_soccer 2019', x_axis_label='Wage', y_axis_label='Value', outer_width=400, outer_height=400, tools=[TOOLTIPS])
# Add circles to the plot
p.circle('Wage', 'Value', size=10, source=df1, color="navy", alpha=0.5)
show(p)
```
